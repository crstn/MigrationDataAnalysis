Not good, the thousands separators threw pandas off. Let's read the file again and tell pandas to ignore the thousands separator:

```python
gdp = pd.read_csv('GDP.csv', sep=";", index_col=0, thousands='.')
print(gdp.dtypes) # use "print" in a jupyter notebook if you want multiple outputs per code cell, like here ﹁
print(gdp) #                                                                                             < - 」
```

Now we have proper numbers, but they are floating point numbers. Not really a problem (also, we can't fix this, since integer data type cannot handle NAs)

How can we get different rows or columns of the GDP dataframe?

```python
gdp.loc["Australia",:]
```

```python
gdp.loc[:,"1995":"1998"]
```

```python
gdp.loc["Germany","2015"]
```

Let's do some simple anaylsis on the dataframe.

Plot the GDP of Syria over time:

```python
gdp.loc["Syrian Arab Republic",:].plot()
```

Let's make this a bit nicer:

```python
syria = gdp.loc["Syrian Arab Republic",:]
afg = gdp.loc["Afghanistan",:]
plt.figure()
plt.title("GDP Syria and Afghanistan", size="x-large")
plt.plot(syria, label="Syria")
plt.plot(afg, label="Afghanistan")
plt.legend(loc=(0.7,0.8))
```

Let's try a scatterplot of GDP versus number of refugees. This is a bit tricky, since we want a data frame with 2 colums as a combined index (country and year) and then two data columns (gdp and migrants). This is done with a [multiindex](https://pandas.pydata.org/pandas-docs/stable/advanced.html) and super easy to generate from the original data frame:

```python
migration_stacked = migration.stack()
migration_stacked
```

Voilà. Now we just do the same to the GDP data

```python
gdp_stacked = gdp.stack()
gdp_stacked
```

Let's rename the columns on the two new data frames:

```python
pd.merge(migration_stacked, gdp_stacked, on='Country Name')
```
