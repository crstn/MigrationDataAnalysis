Let's make this a bit nicer:

Let's try a scatterplot of GDP versus number of refugees. This is a bit tricky, since we want a data frame with 2 colums as a combined index (country and year) and then two data columns (gdp and migrants). This is done with a [multiindex](https://pandas.pydata.org/pandas-docs/stable/advanced.html) and super easy to generate from the original data frame:

```python
migration_stacked = migration.stack()
migration_stacked
```

Voilà. Now we just do the same to the GDP data

```python
gdp_stacked = gdp.stack()
gdp_stacked
```

Let's give the index columns proper names:

```python
gdp_stacked.index.names = ['Country', 'Year']
migration_stacked.index.names = ['Country', 'Year']
```

```python
all_data = pd.merge(migration_stacked.reset_index(), gdp_stacked.reset_index(), on=['Country', 'Year'])
all_data
```

Now we have a plain data frame again. Let's properly index that again on country and year:

```python
all_data.set_index(['Country', 'Year'], inplace=True)
all_data
```

Cool 😎 Now we just need to rename the actual data columns:

```python
all_data.rename(index=str, columns={"0_x": "Migrants", "0_y": "GDP"}, inplace=True)
all_data
```

✨ Nice and clean! Now let's make a scatter plot to see if there's any indication of a relationship between the GDP and number of migrants: 

```python
all_data.plot.scatter(x='GDP', y='Migrants');
```
